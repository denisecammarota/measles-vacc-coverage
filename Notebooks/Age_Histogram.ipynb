```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
```

```python
sys.path.append('..')
```

```python
df_cases = pd.read_csv('../Data/analyzed/age_vaccine_municipalities.csv')
```

```python
df_cases['CASE'] = 1
```

```python
min_age = 0
max_age = int(max(df_cases['NU_IDADE_N']))
plt.hist(df_cases['NU_IDADE_N'], edgecolor = 'black', bins=np.arange(min_age,max_age+1,1))
plt.xticks(np.arange(min_age,max_age+1,5), rotation = 45)
plt.xlabel('Age', fontsize = 12)
plt.ylabel('Cases per 100.000 inhabitants', fontsize = 12)
plt.savefig('../Figs/hist_cases_age.pdf',bbox_inches='tight')
plt.show()
```
